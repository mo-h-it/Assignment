##### Question 1: Some of the orders are stored in another csv file named 'bigkart_newsales'. Read the csv file, store it in a data frame and add it to the 'bigkart_sales' data frame. Find the total sales value of the category 'Office Supplies' after combining the dataframes

```python
##importing the libraries 
import pandas as pd
import numpy as np
```

```python
## Reading the files

df = pd.read_csv(r'C:\Users\Mohit\OneDrive\Desktop\assignment\Assignment For DA Candidates\Data Analyst Assignment Part 2\bigkart_newsales.csv')
df2 = pd.read_csv(r'C:\Users\Mohit\OneDrive\Desktop\assignment\Assignment For DA Candidates\Data Analyst Assignment Part 2\bigkart_sales.csv')
```

```python
dfcomb= pd. concat([df, df2], axis=0).reset_index(drop = True)
```

```python
dfcomb.head()
```

```python
##Total sales of Category office supplies
dfcomb[dfcomb['Category']=='Office Supplies'].groupby(['Category'])['Sales'].sum()
```

### There are some duplicate rows in the data frame. Drop these rows and calculate the total sales value of the category Office Supplies.

```python
## dropping duplicates from whole dataset
dfunique = dfcomb.drop_duplicates()
```

```python
##total sales value of the category Office Supplies after dropping duplicates
dfunique[dfunique['Category']=='Office Supplies'].groupby(['Category'])['Sales'].sum()
```

```python
##Another method
dfcomb[dfcomb['Category']=='Office Supplies'].drop_duplicates().groupby(['Category'])['Sales'].sum().reset_index(drop=True)
```

### Find the most profitable category and sub category combination based on the net profit.

```python
dfunique.groupby(['Category','Sub-Category'], as_index=False)['Profit'].sum().sort_values(by = 'Profit',ascending=False).head(1)
```

#### How many invalid order IDs are there in the data frame. An order id is of the form AZ-2011-Y where Y represents a whole number. A Order ID is said to be valid only if Y consists of 7 digits. Find the number of invalid order order IDs in the data frame.



```python
## Spliting the Order ID
dd = dfunique['Order ID'].str.split('-',expand = True)
dd
```

```python
##Applied a lambda function where length of Y is not equals to 7
wrong_ids =list(filter(lambda e: len(e) !=7, dd[2]))
```

```python
wrong_ids_count = len([i for i in wrong_ids])
```

#### Find the top 25 orders based on sales value and find the number of orders which belong to furniture category.

```python
## Top 25 orders based on sales value
top_25_sales=dfunique.groupby(['Order ID','Category'],as_index = False)['Sales'].sum().sort_values('Sales',ascending = False).head(25).reset_index(drop = True)
```

```python
## Number of orders which belong to furniture category

top_25_sales['Category'][top_25_sales['Category']=='Furniture'].count()
```

#### Among the orders with sales>250 and profit>50, find the product name of the fourth highest order based on sales value.

```python
fourth_high = dfunique[(dfunique['Sales']>250) &  (dfunique['Profit'] > 50)].groupby(['Product Name'],as_index = False)['Sales','Profit'].sum().sort_values('Sales',ascending = False)
```

```python
##Product Name of the fourth highest order based on sales value.
fourth_high.iloc[3][0]
```

#### Remove the orders with negative profit by dropping the corresponding rows with negative `Profit`. Find the product that makes the lowest profit per Quantity in the Technology category.

```python
##Geting the index of negative profits
neg_index = dfunique[dfunique['Profit'] < 0].index
```

```python
dfunique[dfunique['Profit'] < 0]
```

```python
##dropping the negative profit rows
dfunique.drop(neg_index,axis = 0,inplace = True)
```

```python
## Calculating profit per quantity
dfunique['Profit_per_quantity'] = dfunique['Profit']/dfunique['Quantity']
```

```python
dfunique
```

```python
## product that makes the lowest profit per Quantity in the Technology category
dfunique[dfunique['Category']== 'Technology'].groupby('Product Name',as_index = False)['Profit_per_quantity'].sum().min()
```
